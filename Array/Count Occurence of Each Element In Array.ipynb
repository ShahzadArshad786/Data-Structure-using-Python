```python
import numpy as np 

n = int(input("Enter Size of the Array : "))

arr = np.ndarray(shape=(n,) , dtype=int)

print("\nInput ",n, " Elements in the Array\n")

for i in range(n):
    arr[i] = int(input("Input Array Element at index arr[%d] : " %i))

print("\nOriginal Array Elements Are : ", end="")
for i in range(n):
    print(arr[i], end=" ")

print()

arr2 = np.ndarray(shape=(n,) , dtype=int)
visited = -1
for i in range(n):
    count = 1
    for j in range(i + 1, n):
        if arr[i] == arr[j]:
            count = count + 1
            arr2[j] = visited 
   
    if arr2[i] != visited:
        arr2[i] = count 
        
print("---------------------")
print(" Element | Frequency")
print("---------------------")
for i in range(0, len(arr2)):    
    if(arr2[i] != visited):    
        print("  " , arr[i] ,"    |    " , arr2[i])
print("---------------------")
```
