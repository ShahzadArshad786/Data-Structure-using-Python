#### Write a Python program that inputs two matrices and display the product of both matrices. 

```python
import numpy as np 

r = int(input("Enter No. Of Rows    : "))
c = int(input("Enter No. Of Columns : "))

arr1 = np.ndarray(shape = (r,c) , dtype = int) 
arr2 = np.ndarray(shape = (r,c) , dtype = int) 
arr3 = np.ndarray(shape = (r,c) , dtype = int)

print("\nInput 1st Matrix\n")
for i in range(0, r):
    for j in range(0, c):
        arr1[i][j] = int(input("Enter Element at arr1[%d][%d] : " %(i,j)))

        
print("\nInput 2nd Matrix\n")
for i in range(0, r):
    for j in range(0, c):
        arr2[i][j] = int(input("Enter Element at arr2[%d][%d] : " %(i,j)))
        
        
print("\n1st Matrix : \n")
for i in range(0 , r):
    for j in range(0 , c):
        print(arr1[i][j] , "\t", end = " ")
    print("\n")
    
print("\n2nd Matrix : \n")
for i in range(0 , r):
    for j in range(0 , c):
        print(arr1[i][j] , "\t", end = " ")
    print("\n")    

# Product of Two Matrices and Assign the Product values to 3rd Matrix
for i in range(0, r):
    for j in range(0, c):
        arr3[i][j] = 0 
        for k in range(0, c):
            arr3[i][j] = arr3[i][j] + arr1[i][k] * arr2[k][j]
            
print("\nProduct Of Two Matrices : \n")

# Display Product of Two Matrices 
for i in range(0, r):
    for j in range(0, c):
        print(arr3[i][j] , "\t", end = " ")
    print("\n")    
```
