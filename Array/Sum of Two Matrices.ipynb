#### Write a Python program that inputs Two Matrices and Display the Sum of Two Matrices.

```python
import numpy as np 

r = int(input("Enter No. Of Rows    : "))
c = int(input("Enter No. Of Columns : "))

arr1 = np.ndarray(shape = (r,c) , dtype = int) 
arr2 = np.ndarray(shape = (r,c) , dtype = int) 
arr3 = np.ndarray(shape = (r,c) , dtype = int)

print("\nInput 1st Matrix\n")
for i in range(0, r):
    for j in range(0, c):
        arr1[i][j] = int(input("Enter Element at arr1[%d][%d] : " %(i,j)))

        
print("\nInput 2nd Matrix\n")
for i in range(0, r):
    for j in range(0, c):
        arr2[i][j] = int(input("Enter Element at arr2[%d][%d] : " %(i,j)))
        
        
print("\n1st Matrix : \n")
for i in range(0 , r):
    for j in range(0 , c):
        print(arr1[i][j] , "\t", end = " ")
    print("\n")
    
print("\n2nd Matrix : \n")
for i in range(0 , r):
    for j in range(0 , c):
        print(arr1[i][j] , "\t", end = " ")
    print("\n")    

# Sum of Two Matrices and Assign the sum values to 3rd Matrix
for i in range(0, r):
    for j in range(0, c):
        arr3[i][j] = arr1[i][j] + arr2[i][j]
            
print("\nSum Of Two Matrices : \n")

# Display Sum of Two Matrices 
for i in range(0, r):
    for j in range(0, c):
        print(arr3[i][j] , "\t", end = " ")
    print("\n")    
```
